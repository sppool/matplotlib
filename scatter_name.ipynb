```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['font.sans-serif'] = ['Gen Jyuu Gothic P'] # 中文輸出(源柔黑體)
```

```python
def fig_text(text='HELLO', N=1000, rseed=0):
    # Make a plot with "HELLO" text; save as PNG
    fig, ax = plt.subplots(figsize=(4, 1))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    ax.text(0.5, 0.4, text, va='center', ha='center', weight='bold', size=60)
    fig.savefig('text.png')
    plt.close(fig)
    
    # Open this PNG and draw random points from it
    from matplotlib.image import imread
    data = imread('text.png')[::-1, :, 0].T
    rng = np.random.RandomState(rseed)
    X = rng.rand(4 * N, 2)
    i, j = (X * data.shape).astype(int).T
    mask = (data[i, j] < 1)
    X = X[mask]
    X[:, 0] *= (data.shape[0] / data.shape[1])
    X = X[:N]
    return X[np.argsort(X[:, 0])]
```

```python
X = fig_text('Allison', 5000)
colorize = dict(c=X[:, 1], s=2, cmap=matplotlib.cm.spring)
plt.scatter(X[:, 0], X[:, 1], **colorize)
plt.axis('equal');
```

```python
X = fig_text('Allison', 5000)
colorize = dict(c=X[:, 0] + (2 * X[:, 1]), s=2, cmap=matplotlib.cm.cool_r)
plt.scatter(X[:, 0], X[:, 1], **colorize)
plt.axis('equal');
```

```python
X = fig_text('尤意茹', 5000)
colorize = dict(c=X[:, 0]+X[:, 1], s=3, cmap=matplotlib.cm.autumn)
plt.scatter(X[:, 0], X[:, 1], **colorize)
plt.axis('equal');
```

```python
X = fig_text('尤意茹', 10000)
colorize = dict(c=X[:, 0] + (3 * X[:, 1]), s=1, cmap=matplotlib.cm.spring_r)
plt.scatter(X[:, 0], X[:, 1], **colorize)
plt.axis('equal');
```
